```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Review = pd.read_csv(r"WordCloudData.csv", encoding ="latin-1")
```

```python
# Iterating through the .csv data file 
comment_words = ''
for i in Review.ReviewText: 
    i = str(i) 
    separate = i.split() 
    for j in range(len(separate)): 
        separate[j] = separate[j].lower() 
      
    comment_words += " ".join(separate)+" "
```

```python
stop_words = ["a", "an", "the", "this", "that", "is", "it", "to", "and","i","am","have","as","in","for","on","with","but","her","him","also","from","just","had","of","if"]
final_wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stop_words, 
                min_font_size = 10).generate(comment_words)
```

```python
# Displaying the WordCloud                    
plt.figure(figsize = (6, 8), facecolor = None) 
plt.imshow(final_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.title("Most used words in reviews", size=16)   
plt.show()
```
